```python
print('BTC')
BTC.drop(['High', 'Low','Open', 'Close', 'Volume'], axis='columns', inplace=True)
BTC['Adj Close'] = BTC['Adj Close'] / BTC.loc['2014-01-01']['Adj Close']
BTC.plot()
```

```python
print('LTC')
LTC.drop(['High', 'Low','Open', 'Close', 'Volume'], axis='columns', inplace=True)
LTC['Adj Close'] = LTC['Adj Close'] / LTC.loc['2014-01-01']['Adj Close']
LTC.plot()
```

On all the graphs we see a big peak which matches with the speculative bubble of Jannuary 2018. Before this the cryptocurrencies were growing and they seem to growing since the big correction of 2018. There seems to be a strong correlation between them
