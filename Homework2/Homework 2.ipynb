<h2>Question6</h2>

```python
from scipy.interpolate import interp1d
from math import *
```

```python
df30days = dfVolatility.loc[dfVolatility['maturity']==30,:]
df30days
```

```python
xdate = df30days['date']
x=[]
for i in xdate:
    date = i -20160000
    mois = floor(date /100)
    jour = date - mois *100
    if(mois == 1):
        x.append(jour)
    elif(mois == 2):
        x.append(jour+31)
    elif(mois == 3):
        x.append(jour+60)
    else:
        x.append(jour+91)
y = df30days['volatility']
f = interp1d(x, y)
f2 = interp1d(x, y, kind='cubic')
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, 'o', x, f(x), '-', x, f2(x), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.xlabel('Days')
plt.ylabel('Volatility')
plt.show()
```
